# Scaling laws

```python
import numpy as N
import matplotlib.pyplot as P
import pandas as pd
P.style.use(['seaborn-white', 'seaborn-paper', 'seaborn-ticks'])
matplotlib.rc("font", family="Times New Roman", size=12)
matplotlib.rc('xtick', labelsize='medium')
matplotlib.rc('ytick', labelsize='medium')
```

```python
sims1= pd.read_csv('./tables/data_results_nrho1.txt', sep=",")
sims1['color'] = sims1['eta'].replace(0.4,'k').replace(0.6,'r')
sims1['symbol'] = sims1['Pr'].replace(1,'d')
```

```python
sims15= pd.read_csv('./tables/data_results_nrho15.txt', sep=",")
sims15['color'] = sims15['eta'].replace(0.4,'k').replace(0.6,'r')
sims15['symbol'] = sims15['Pr'].replace(1,'v')
```

```python
sims3= pd.read_csv('./tables/data_results_nrho3.txt', sep=",")
sims3['color'] = sims1['eta'].replace(0.4,'k').replace(0.6,'r')
sims3['symbol'] = sims1['Pr'].replace(1,'p')
```

```python
P.scatter(sims3['Rac'],  sims3['Ro'],  c=sims3['color'],  marker='p', s=80)
P.scatter(sims15['Rac'], sims15['Ro'], c=sims15['color'], marker='v', s=80)
P.scatter(sims1['Rac'],  sims1['Ro'],  c=sims1['color'],  marker='d', s=80)
P.yscale('log')
P.ylabel('Ro', fontsize='large')
P.xlabel(r'Ra (Ra$_\mathrm{c}$)', fontsize='large')
P.title('Rossby vs Rayleigh', fontsize='large')
```

```python
P.scatter(sims3['Ro'], sims3['Ekin'], c=sims3['color'], marker='p', s=80)
P.scatter(sims15['Ro'], sims15['Ekin'], c=sims15['color'], marker='v', s=80)
P.scatter(sims1['Ro'], sims1['Ekin'], c=sims1['color'], marker='d', s=80)
P.yscale('log')
P.xlabel('Ro', fontsize='large')
P.ylabel(r'$E_\mathrm{K}$', fontsize='large')
P.title('Kinetic energy', fontsize='large')
```

```python
P.scatter(sims3['Ro'],  sims3['Emp_ax']/sims3['Emp'],   c=sims3['color'],  marker='p', s=80)
P.scatter(sims15['Ro'], sims15['Emp_ax']/sims15['Emp'], c=sims15['color'], marker='v', s=80)
P.scatter(sims1['Ro'],  sims1['Emp_ax']/sims1['Emp'],   c=sims1['color'],  marker='d', s=80)
P.xscale('log')
P.xlabel('Ro', fontsize='large')
P.ylabel(r'$E_\mathrm{P, axi} \; (E_\mathrm{P})$', fontsize='large')
P.title('Axisymmetric magnetic energy', fontsize='large')
```

```python
P.scatter(sims3['Ro'],  sims3['Emp_ax']/sims3['Emag'],   c=sims3['color'],  marker='p', s=80)
P.scatter(sims15['Ro'], sims15['Emp_ax']/sims15['Emag'], c=sims15['color'], marker='v', s=80)
P.scatter(sims1['Ro'],  sims1['Emp_ax']/sims1['Emag'],   c=sims1['color'],  marker='d', s=80)
P.xscale('log')
P.xlabel('Ro', fontsize='large')
P.ylabel(r'$E_\mathrm{P, axi} \; (E_\mathrm{M})$', fontsize='large')
P.title('Axisymmetric magnetic energy', fontsize='large')
```

```python
P.scatter(sims3['Ro'],  sims3['Emp']/sims3['Emag'],   c=sims3['color'],  marker='p', s=80)
P.scatter(sims15['Ro'], sims15['Emp']/sims15['Emag'], c=sims15['color'], marker='v', s=80)
P.scatter(sims1['Ro'],  sims1['Emp']/sims1['Emag'],   c=sims1['color'],  marker='d', s=80)
#P.xscale('log')
P.xlabel('Ro', fontsize='large')
P.ylabel(r'$E_\mathrm{P} \; (E_\mathrm{M})$', fontsize='large')
P.title('Poloidal magnetic energy', fontsize='large')
```

```python
P.scatter(sims3['Rac'],  sims3['Emp']/sims3['Emag'],   c=sims3['color'],  marker='p', s=80)
P.scatter(sims15['Rac'], sims15['Emp']/sims15['Emag'], c=sims15['color'], marker='v', s=80)
P.scatter(sims1['Rac'],  sims1['Emp']/sims1['Emag'],   c=sims1['color'],  marker='d', s=80)
P.xlabel('Ra (Ra$_\mathrm{c}$)', fontsize='large')
P.ylabel(r'$E_\mathrm{P} \; (E_\mathrm{M})$', fontsize='large')
P.title('Poloidal magnetic energy', fontsize='large')
```
