# HD simulations

```python
# Define paths of simulations to be compared
WORK_DIR = "/Volumes/LaCie/Simulations/hd/dif_eta/"
PATH1 = WORK_DIR + "eta20Ek5Ra8dS100"          # eta 20
PATH2 = WORK_DIR + "eta40Ek5Ra8dS100"          # eta 40
PATH3 = WORK_DIR + "eta50Ek5Ra8dS100"          # eta 50
PATH4 = WORK_DIR + "eta60Ek5Ra8dS100"          # eta 60
```

```python
# Read simulations' data
os.chdir(PATH1)
ts1 = MagicTs(field='par', all=True, iplot=False)
os.chdir(PATH2)
ts2 = MagicTs(field='par', all=True, iplot=False)
os.chdir(PATH3)
ts3 = MagicTs(field='par', all=True, iplot=False)
os.chdir(PATH4)
ts4 = MagicTs(field='par', all=True, iplot=False)
```

```python
### RESCALING Ro TO MATCH WITH THE DEFINITION IN GASTINE ET AL. 2012
scale=0.4 # size of the convective zone
it1 = ts1.time.argsort()
it2 = ts2.time.argsort()
it3 = ts3.time.argsort()
it4 = ts4.time.argsort()
plt.plot(ts1.time[it1], ts1.rolc[it1]/scale, 'k', label=r'$\eta=0.2$', alpha=0.8)
plt.plot(ts2.time[it2], ts2.rolc[it2]/scale, 'r', label=r'$\eta=0.4$', alpha=0.7)
plt.plot(ts3.time[it3], ts3.rolc[it3]/scale, 'b', label=r'$\eta=0.5$', alpha=0.6)
plt.plot(ts4.time[it4], ts4.rolc[it4]/scale, 'g', label=r'$\eta=0.6$', alpha=0.3)
plt.ylabel(r'$Ro_\mathrm{l}$', fontsize=20)
plt.xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
plt.ylim((0.025,.17))
plt.xlim((0.0,.2))
#plt.yscale('log')
plt.legend(fontsize=15)
plt.tight_layout()
plt.show()
```

```python
### RESCALING Ro TO MATCH WITH THE DEFINITION IN GASTINE ET AL. 2012
scale=0.4 # size of the convective zone
it1 = ts1.time.argsort()
it2 = ts2.time.argsort()
it3 = ts3.time.argsort()
it4 = ts4.time.argsort()
plt.plot(ts1.time[it1], ts1.rm[it1]*scale, 'k', label=r'$\eta=0.2$', alpha=0.8)
plt.plot(ts2.time[it2], ts2.rm[it2]*scale, 'r', label=r'$\eta=0.4$', alpha=0.7)
plt.plot(ts3.time[it3], ts3.rm[it3]*scale, 'b', label=r'$\eta=0.5$', alpha=0.6)
plt.plot(ts4.time[it4], ts4.rm[it4]*scale, 'g', label=r'$\eta=0.6$', alpha=0.3)
plt.ylabel(r'$Rm$', fontsize=20)
plt.xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
#plt.ylim((0.025,.17))
plt.xlim((0.0,.2))
#plt.yscale('log')
plt.legend(fontsize=15)
plt.tight_layout()
plt.show()
```
