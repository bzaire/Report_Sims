```python
#Paths for simulations and for the folder to store figures
FIGS_DIR = "/Volumes/LaCie/Report_Sims/figs/"
WORK_DIR = "/Volumes/LaCie/Simulations/hd/dif_eta/"

PATH1 = WORK_DIR + "eta20Ek5Ra8dS100"          # eta 20
PATH2 = WORK_DIR + "eta40Ek5Ra8dS100"          # eta 40
PATH3 = WORK_DIR + "eta50Ek5Ra8dS100"          # eta 50
PATH4 = WORK_DIR + "eta60Ek5Ra8dS100"          # eta 60

# Name of figures
FIGNAME1 = 'HD_equat_eta'
```

```python
# Read simulations' data
os.chdir(PATH1)
ts1 = MagicTs(field='par', all=True, iplot=False)
os.chdir(PATH2)
ts2 = MagicTs(field='par', all=True, iplot=False)
os.chdir(PATH3)
ts3 = MagicTs(field='par', all=True, iplot=False)
os.chdir(PATH4)
ts4 = MagicTs(field='par', all=True, iplot=False)
```

```python
### RESCALING Rm TO MATCH WITH THE DEFINITION IN GASTINE ET AL. 2012
scale=0.4 # size of the convective zone
it1 = ts1.time.argsort()
it2 = ts2.time.argsort()
it3 = ts3.time.argsort()
it4 = ts4.time.argsort()
plt.plot(ts1.time[it1], ts1.rm[it1]*scale, 'k', label=r'$\eta=0.2$', alpha=0.8)
plt.plot(ts2.time[it2], ts2.rm[it2]*scale, 'r', label=r'$\eta=0.4$', alpha=0.7)
plt.plot(ts3.time[it3], ts3.rm[it3]*scale, 'b', label=r'$\eta=0.5$', alpha=0.6)
plt.plot(ts4.time[it4], ts4.rm[it4]*scale, 'g', label=r'$\eta=0.6$', alpha=0.3)
plt.ylabel(r'$Rm$', fontsize=20)
plt.xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
#plt.ylim((0.025,.17))
plt.xlim((0.0,.2))
#plt.yscale('log')
plt.legend(fontsize=15)
plt.tight_layout()
plt.show()
```

```python
### RESCALING Ro TO MATCH WITH THE DEFINITION IN GASTINE ET AL. 2012
scale=0.4 # size of the convective zone
it1 = ts1.time.argsort()
it2 = ts2.time.argsort()
it3 = ts3.time.argsort()
it4 = ts4.time.argsort()
plt.plot(ts1.time[it1], ts1.rolc[it1]/scale, 'k', label=r'$\eta=0.2$', alpha=0.8)
plt.plot(ts2.time[it2], ts2.rolc[it2]/scale, 'r', label=r'$\eta=0.4$', alpha=0.7)
plt.plot(ts3.time[it3], ts3.rolc[it3]/scale, 'b', label=r'$\eta=0.5$', alpha=0.6)
plt.plot(ts4.time[it4], ts4.rolc[it4]/scale, 'g', label=r'$\eta=0.6$', alpha=0.3)
plt.ylabel(r'$Ro_\mathrm{l}$', fontsize=20)
plt.xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
plt.ylim((0.025,.17))
plt.xlim((0.0,.2))
#plt.yscale('log')
plt.legend(fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# Read simulations' data
os.chdir(PATH1)
rad1 = MagicRadial(field='parR', iplot=False)
os.chdir(PATH2)
rad2 = MagicRadial(field='parR', iplot=False)
os.chdir(PATH3)
rad3 = MagicRadial(field='parR', iplot=False)
os.chdir(PATH4)
rad4 = MagicRadial(field='parR', iplot=False)
```

```python
### RESCALING Ro TO MATCH WITH THE DEFINITION IN GASTINE ET AL. 2012
scale=0.4 # size of the convective zone
plt.plot(rad1.radius, rad1.rol/scale, 'k', label=r'$\eta=0.2$', alpha=0.8)
plt.plot(rad2.radius, rad2.rol/scale, 'r', label=r'$\eta=0.4$', alpha=0.8)
plt.plot(rad3.radius, rad3.rol/scale, 'b', label=r'$\eta=0.5$', alpha=0.8)
plt.plot(rad4.radius, rad4.rol/scale, 'g', label=r'$\eta=0.6$', alpha=0.8)
plt.ylabel(r'$Ro_\mathrm{l}$', fontsize=20)
plt.xlabel(r'$r \; (r_\mathrm{o})$', fontsize=20)
plt.legend(fontsize=15)
plt.tight_layout()
plt.show()
```

```python
#Local flow length-scale (based on the mass-weighted velocity)
fig, ax = plt.subplots(1,2, figsize=(10,5), gridspec_kw={'wspace': 0.3})
ax[0].plot(rad1.radius, rad1.dlV, 'k', label=r'$\eta=0.2$', alpha=0.8)
ax[0].plot(rad2.radius, rad2.dlV, 'r', label=r'$\eta=0.4$', alpha=0.8)
ax[0].plot(rad3.radius, rad3.dlV, 'b', label=r'$\eta=0.5$', alpha=0.8)
ax[0].plot(rad4.radius, rad4.dlV, 'g', label=r'$\eta=0.6$', alpha=0.8)
ax[0].set_ylabel(r'$L$', fontsize=20)
ax[0].set_xlabel(r'$r \; (r_\mathrm{o})$', fontsize=20)
ax[0].set_xticks((0.6,0.8,1.))
ax[0].set_xlim((0.4, 1.))
ax[0].legend(fontsize=15)


#Local flow length-scale based on the non-axisymmetric flow components (based on the RMS velocity)
ax[1].plot(rad1.radius, rad1.dlVc, 'k', label=r'$\eta=0.2$', alpha=0.8)
ax[1].plot(rad2.radius, rad2.dlVc, 'r', label=r'$\eta=0.4$', alpha=0.8)
ax[1].plot(rad3.radius, rad3.dlVc, 'b', label=r'$\eta=0.5$', alpha=0.8)
ax[1].plot(rad4.radius, rad4.dlVc, 'g', label=r'$\eta=0.6$', alpha=0.8)
ax[1].set_ylabel(r'$L$', fontsize=20)
ax[1].set_xlabel(r'$r \; (r_\mathrm{o})$', fontsize=20)
ax[1].set_title('non-Axi', fontsize=20)
ax[1].set_xlim((0.4, 1.))
#ax[1].set_ylim((0.05, .30))
#ax[1].legend(fontsize=15)
```

```python
# Read simulations' data
os.chdir(PATH1)
ts1 = MagicTs(field='e_kin', all=True, iplot=False)
os.chdir(PATH2)
ts2 = MagicTs(field='e_kin', all=True, iplot=False)
os.chdir(PATH3)
ts3 = MagicTs(field='e_kin', all=True, iplot=False)
os.chdir(PATH4)
ts4 = MagicTs(field='e_kin', all=True, iplot=False)
```

```python
### RESCALING Ekin TO MATCH WITH THE DEFINITION IN GASTINE ET AL. 2012
scale=0.4 # size of the convective zone
it1 = ts1.time.argsort()
it2 = ts2.time.argsort()
it3 = ts3.time.argsort()
it4 = ts4.time.argsort()
plt.plot(ts1.time[it1], ts1.ekin_tot[it1]*scale**2, 'k', label=r'$\eta=0.2$', alpha=0.8)
plt.plot(ts2.time[it2], ts2.ekin_tot[it2]*scale**2, 'r', label=r'$\eta=0.4$', alpha=0.7)
plt.plot(ts3.time[it3], ts3.ekin_tot[it3]*scale**2, 'b', label=r'$\eta=0.5$', alpha=0.6)
plt.plot(ts4.time[it4], ts4.ekin_tot[it4]*scale**2, 'g', label=r'$\eta=0.6$', alpha=0.3)
plt.ylabel(r'$E_\mathrm{k}$', fontsize=20)
plt.xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
plt.ylim((8.e4,4.e5))
plt.xlim((0.0,.2))
plt.yscale('log')
plt.legend(fontsize=15)
plt.tight_layout()
plt.show()
```
