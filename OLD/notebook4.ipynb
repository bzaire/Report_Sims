# Magnetic results
### BC: constant entropy

Here we look for the results on simualtions with constant entropy as the boundary condition. 
We compare FC and PC simulations at different criticalities. 

```python
import os 
from magic import *

#Paths for simulations and for the folder to store figures
FIGS_DIR = "/Volumes/LaCie/Report_Sims/figs/"
WORK_DIR = "/Volumes/LaCie/Simulations/mhd/"

# Sims w/ constant entropy at the boundaries:
## FC
FC1 = WORK_DIR + 'Nr15Ek5Ra7dS328Pm5-FC_bs'
FC2 = WORK_DIR + 'Nr15Ek5Ra7dS250Pm5-FC_bs'
FC3 = WORK_DIR + 'Nr15Ek5Ra7dS200Pm5-FC_bs'
## PC
PC1 = WORK_DIR + 'Nr15Ek5Ra7dS328Pm5_bs'
PC2 = WORK_DIR + 'Nr15Ek5Ra7dS250Pm5_bs'
PC3 = WORK_DIR + 'Nr15Ek5Ra7dS200Pm5_bs'

# Name of figures
FIGNAME1 = 'FC_fdip'
FIGNAME2 = 'PC_fdip'
FIGNAME3 = 'FC_surf_'
FIGNAME4 = 'PC_surf_'
FIGNAME5 = 'FC_equat_'
FIGNAME6 = 'PC_equat_'
```

## Fully convective simulations

```python
# Rescaling to match Gastine et al. 2012, their scale of reference is not the outer radius but the convective shell gap. 
scale = 0.4

os.chdir(FC1)
ts1 = MagicTs(field='dipole', all=True, iplot=False)
rad1 = MagicRadial(field='parR', iplot=False)
it1 = ts1.time.argsort()

os.chdir(FC2)
ts2 = MagicTs(field='dipole', all=True, iplot=False)
rad2 = MagicRadial(field='parR', iplot=False)
it2 = ts2.time.argsort()


os.chdir(FC3)
ts3 = MagicTs(field='dipole', all=True, iplot=False)
rad3 = MagicRadial(field='parR', iplot=False)
it3 = ts3.time.argsort()


os.chdir(FIGS_DIR)
fig, ax = plt.subplots(1,3, sharey=True, figsize=(15,5))
ax[0].plot(ts1.time[it1], ts1.dip_cmb[it1], 'k', label='CMB')
ax[0].plot(ts1.time[it1], ts1.dipolarity[it1], 'r', label='Total')
ax[0].set_ylabel(r'$f_\mathrm{dip}$', fontsize=20)
ax[0].set_xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
ax[0].set_title(r'$Ra=1.5 Ra_\mathrm{c}$', fontsize=20)
ax[0].legend(fontsize=15)


ax[1].plot(ts2.time[it2], ts2.dip_cmb[it2], 'k', label='CMB')
ax[1].plot(ts2.time[it2], ts2.dipolarity[it2], 'r', label='Total')
ax[1].set_xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
ax[1].set_title(r'$Ra=2.0 Ra_\mathrm{c}$', fontsize=20)
ax[1].legend(fontsize=15)


# FC results
ax[2].plot(ts3.time[it3], ts3.dip_cmb[it3], 'k', label='CMB')
ax[2].plot(ts3.time[it3], ts3.dipolarity[it3], 'r', label='Total')
ax[2].set_xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
ax[2].set_title(r'$Ra=2.5 Ra_\mathrm{c}$', fontsize=20)
ax[2].legend(fontsize=15)
plt.tight_layout()
plt.savefig(FIGNAME1+'.png')
plt.show()
```

## Partly convective simulations


```python
# Rescaling to match Gastine et al. 2012, their scale of reference is not the outer radius but the convective shell gap. 
scale = 0.4

os.chdir(PC1)
ts1_pc = MagicTs(field='dipole', all=True, iplot=False)
rad1_pc = MagicRadial(field='parR', iplot=False)
it1_pc = ts1_pc.time.argsort()

os.chdir(PC2)
ts2_pc = MagicTs(field='dipole', all=True, iplot=False)
rad2_pc = MagicRadial(field='parR', iplot=False)
it2_pc = ts2_pc.time.argsort()


os.chdir(PC3)
ts3_pc = MagicTs(field='dipole', all=True, iplot=False)
rad3_pc = MagicRadial(field='parR', iplot=False)
it3_pc = ts3_pc.time.argsort()


os.chdir(FIGS_DIR)
fig, ax = plt.subplots(1,3, sharey=True, figsize=(15,5))
ax[0].plot(ts1_pc.time[it1_pc], ts1_pc.dip_cmb[it1_pc], 'k', label='CMB')
ax[0].plot(ts1_pc.time[it1_pc], ts1_pc.dipolarity[it1_pc], 'r', label='Total')
ax[0].set_ylabel(r'$f_\mathrm{dip}$', fontsize=20)
ax[0].set_xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
ax[0].set_title(r'$Ra=1.5 Ra_\mathrm{c}$', fontsize=20)
ax[0].legend(fontsize=15)


ax[1].plot(ts2_pc.time[it2_pc], ts2_pc.dip_cmb[it2_pc], 'k', label='CMB')
ax[1].plot(ts2_pc.time[it2_pc], ts2_pc.dipolarity[it2_pc], 'r', label='Total')
ax[1].set_xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
ax[1].set_title(r'$Ra=2.0 Ra_\mathrm{c}$', fontsize=20)
ax[1].legend(fontsize=15)


# FC results
ax[2].plot(ts3_pc.time[it3_pc], ts3_pc.dip_cmb[it3_pc], 'k', label='CMB')
ax[2].plot(ts3_pc.time[it3_pc], ts3_pc.dipolarity[it3_pc], 'r', label='Total')
ax[2].set_xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
ax[2].set_title(r'$Ra=2.5 Ra_\mathrm{c}$', fontsize=20)
ax[2].legend(fontsize=15)
plt.tight_layout()
plt.savefig(FIGNAME2+'.png')
plt.show()
```

# Size of the eddies in this simulations

_Continuous lines represent PC simulations and dashed lines the FC ones._

```python
os.chdir(FIGS_DIR)
fig, ax = plt.subplots(1,2, figsize=(15,5))
# FC results
ax[0].plot(rad1.radius, rad1.dlV, '--k', label='1.5Rac')
ax[0].plot(rad2.radius, rad2.dlV, '--r', label='2.0Rac')
ax[0].plot(rad3.radius, rad3.dlV, '--b', label='2.5Rac')
# PC results
ax[0].plot(rad1_pc.radius, rad1_pc.udlVc, 'k')
ax[0].plot(rad2_pc.radius, rad2_pc.udlVc, 'r')
ax[0].plot(rad3_pc.radius, rad3_pc.udlVc, 'b')
ax[0].set_ylabel(r'$L$', fontsize=20)
ax[0].set_xlabel(r'$r \; (r_\mathrm{o})$', fontsize=20)
ax[0].legend(fontsize=15)
# FC results
ax[1].plot(rad1.radius, rad1.dlV, '--k', label='1.5Rac')
ax[1].plot(rad2.radius, rad2.dlV, '--r', label='2.0Rac')
ax[1].plot(rad3.radius, rad3.dlV, '--b', label='2.5Rac')
# PC results
#ax[1].plot(rad1_pc.radius, rad1_pc.udlVc, 'k')
ax[1].plot(rad2_pc.radius, rad2_pc.udlVc, 'r')
ax[1].plot(rad3_pc.radius, rad3_pc.udlVc, 'b')
ax[1].set_ylabel(r'$L$', fontsize=20)
ax[1].set_xlabel(r'$r \; (r_\mathrm{o})$', fontsize=20)
ax[1].set_title('zoom CZ', fontsize=20)
ax[1].set_xlim((0.6, 1.))
ax[1].set_ylim((0.05, .15))
ax[1].legend(fontsize=15)
```

We observe a decrease in the Rossby number in the PC simulations when compared to the FC cases. However, whereas in cases $2.0 Ra_\mathrm{c}$ and $2.5 Ra_\mathrm{c}$ the size of the eddies decreased to respect to the FC analog, in case with $1.5 Ra_\mathrm{c}$ and increase in the eddy size was associated with the decrease in the Rossby number.
