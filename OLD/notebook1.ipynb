# Test of resolution

Here we compare three simulations with different $N_\phi$ points to access the robustness of our solutions. 

```python
# Define paths of simulations to be compared
WORK_DIR = "/Volumes/LaCie/Simulations/mhd/"
PATH1 = WORK_DIR + "Nr15Ek5Ra7dS200Pm5-FC_bs_init0"         # Simulation with Nphi = 360
PATH2 = WORK_DIR + "Nr15Ek5Ra7dS200Pm5-FC_bs_init0_Nphi"    # Simulation with Nphi = 384
PATH3 = WORK_DIR + "Nr15Ek5Ra7dS200Pm5-FC_bs_init0_Nphi512" # Simulation with Nphi = 512
```

## Dipolar fraction

```python
# Read simulations' data
os.chdir(PATH1)
ts1 = MagicTs(field='par', all=True, iplot=False)
os.chdir(PATH2)
ts2 = MagicTs(field='par', all=True, iplot=False)
os.chdir(PATH3)
ts3 = MagicTs(field='par', all=True, iplot=False)
```

```python
it1 = ts1.time.argsort()
it2 = ts2.time.argsort()
it3 = ts3.time.argsort()
plt.plot(ts1.time[it1], ts1.dip_cmb[it1], 'k', label=r'$N_\phi = 360$')
plt.plot(ts2.time[it2], ts2.dip_cmb[it2], 'r', label=r'$N_\phi = 384$')
plt.plot(ts3.time[it3], ts3.dip_cmb[it3], 'b', label=r'$N_\phi = 512$')
plt.ylabel(r'$f_\mathrm{dip}$', fontsize=20)
plt.xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
plt.ylim((0.,1.01))
plt.xlim((0.,0.035))
plt.legend(fontsize=15)
plt.tight_layout()
plt.show()
```

## Rossby and Magnetic Reynolds numbers

```python
fig, ax = plt.subplots(2, sharex=True, figsize=(8, 8))
it1 = ts1.time.argsort()
it2 = ts2.time.argsort()
it3 = ts3.time.argsort()
ax[0].plot(ts1.time[it1], ts1.rm[it1], 'k', label='Nphi = 360')
ax[0].plot(ts2.time[it2], ts2.rm[it2], 'r', label='Nphi = 384')
ax[0].plot(ts3.time[it3], ts3.rm[it3], 'b', label='Nphi = 512')
ax[1].plot(ts1.time[it1], ts1.rolc[it1], 'k')
ax[1].plot(ts2.time[it2], ts2.rolc[it2], 'r')
ax[1].plot(ts3.time[it3], ts3.rolc[it3], 'b')
ax[0].set_ylabel(r'$Rm$', fontsize=20)
ax[1].set_ylabel(r'$Ro_\mathrm{l}$', fontsize=20)
plt.xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
#ax[0].set_yscale('log')
ax[0].set_ylim((600,1200))
#ax[1].set_yscale('log')
ax[1].set_ylim((0.025,0.04))
plt.xlim((0.,0.035))
ax[0].legend(fontsize=15, ncol=1, loc='upper right')
plt.tight_layout()
plt.show()
```

## Total magnetic energy

```python
# Read simulations' data
os.chdir(PATH1)
ts1 = MagicTs(field='e_mag_oc', all=True, iplot=False)
os.chdir(PATH2)
ts2 = MagicTs(field='e_mag_oc', all=True, iplot=False)
os.chdir(PATH3)
ts3 = MagicTs(field='e_mag_oc', all=True, iplot=False)
```

```python
it1 = ts1.time.argsort()
it2 = ts2.time.argsort()
it3 = ts3.time.argsort()
fig = plt.figure(figsize=(7, 4))
plt.plot(ts1.time[it1], ts1.emag_tot[it1], 'k', label='Nphi = 360')
plt.plot(ts2.time[it2], ts2.emag_tot[it2], 'r', label='Nphi = 384')
plt.plot(ts3.time[it3], ts3.emag_tot[it3], 'b', label='Nphi = 512')
plt.ylabel(r'$e_\mathbf{B}$', fontsize=20)
plt.xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
plt.yscale('log')
plt.xlim((0.,0.015))
plt.legend(fontsize=15, loc='lower right')
plt.tight_layout()
plt.show()
```
