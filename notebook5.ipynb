# Magnetic results
### BC: derivative of entropy

Here we look for the results on simualtions in which the boundary condition was zero radial derivative in the entropy fluctuations. 
We compare FC and PC simulations at different criticalities. 

```python
import os 
from magic import *

#Paths for simulations and for the folder to store figures
FIGS_DIR = "/Volumes/LaCie/Report_Sims/figs/"
WORK_DIR = "/Volumes/LaCie/Simulations/mhd/"

# Sims w/ constant entropy at the boundaries:
## FC
FC1 = WORK_DIR + 'Nr15Ek5Ra7dS328Pm5-FC_bdsdr_init0'
FC2 = WORK_DIR + 'Nr15Ek5Ra7dS250Pm5-FC_bdsdr_init0'
FC3 = WORK_DIR + 'Nr15Ek5Ra7dS200Pm5-FC_bdsdr_init0'
## PC
PC1 = WORK_DIR + 'Nr15Ek5Ra7dS328Pm5_bdsdr_init0'
PC2 = WORK_DIR + 'Nr15Ek5Ra7dS250Pm5_bdsdr_init0'
PC3 = WORK_DIR + 'Nr15Ek5Ra7dS200Pm5_bdsdr_init0'

# Name of figures
FIGNAME1 = 'FC_'
FIGNAME2 = 'PC_'
```

## Fully convective simulations

- $Ra = 1.5 Ra_\mathrm{c}$ 

```python
os.chdir(FC1)
ts1 = MagicTs(field='dipole', all=True, iplot=False)
rad1 = MagicRadial(field='parR', iplot=False)
os.chdir(FIGS_DIR)
it1 = ts1.time.argsort()
plt.plot(ts1.time[it1], ts1.dip_cmb[it1], 'k', label='CMB')
plt.plot(ts1.time[it1], ts1.dipolarity[it1], 'r', label='Total')
plt.ylabel(r'$f_\mathrm{dip}$', fontsize=20)
plt.xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
plt.ylim((0.,1.))
plt.legend(fontsize=15)
plt.title(r'$Ra=1.5 Ra_\mathrm{c}$')
plt.tight_layout()
plt.savefig(FIGNAME1+'1_5rac.png')
plt.show()
```

- $Ra = 2.0 Ra_\mathrm{c}$ 

```python
os.chdir(FC2)
ts2 = MagicTs(field='dipole', all=True, iplot=False)
rad2 = MagicRadial(field='parR', iplot=False)
os.chdir(FIGS_DIR)
it2 = ts2.time.argsort()
plt.plot(ts2.time[it2], ts2.dip_cmb[it2], 'k', label='CMB')
plt.plot(ts2.time[it2], ts2.dipolarity[it2], 'r', label='Total')
plt.ylabel(r'$f_\mathrm{dip}$', fontsize=20)
plt.xlabel(r'$t \; (\tau_\nu)$', fontsize=20)
plt.ylim((0.,1.))
plt.title(r'$Ra=2.0 Ra_\mathrm{c}$')
plt.legend(fontsize=15)
plt.tight_layout()
plt.savefig(FIGNAME1+'2_0rac.png')
plt.show()
```
