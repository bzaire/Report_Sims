# Source terms of magnetic field
The mean-field induction equation that discribes the evolution of large-scale magnetic fields is given by 
\begin{equation} 
\frac{d\bar{\mathbf{B}}}{dt} =  \nabla \times (\mathbf{\bar{v}} \times \bar{\mathbf{B}}) + \nabla \times {\mathbf{\mathcal{E}}} - \nabla \times (\eta \nabla \times {\bar{\mathbf{B}}}),
\end{equation}
where $\mathbf{\mathcal{E}} \equiv \overline{\mathbf{v'} \times \mathbf{B'}}$ results from correlations between fluctuating motions and magnetic field. This term physically acts as an electromotive force (EMF). 

If we split the mean-field dynamo equation in its toroidal and poloidal counterparts, we obtain respectively
\begin{equation}
\frac{d\bar{B}_\phi}{dt} =  \left(\bar{\mathbf{B}}_\mathrm{P} . \nabla \right)\bar{v}_\phi  + \left( \nabla \times \mathcal{E} \right)_\phi + \eta \nabla^2{\bar{B}_\phi} \nonumber 
\end{equation}
and
\begin{equation}
\frac{d\bar{A}}{dt} =  
\mathcal{E}_\phi + \eta\nabla^2{\bar{A}}.
\end{equation}

# Source terms with First-order smoothing approximation (FOSA)

## $\alpha$-effect
\begin{equation}
\alpha = - \frac{\tau}{3} \mathcal{H} = - \frac{\tau}{3} (\mathcal{H}_\mathrm{K} + \mathcal{H}_\mathrm{M}),
\end{equation}
with $$\mathcal{H}_\mathrm{K} = \langle u' . \nabla \times u' \rangle_\phi$$ and $$\mathcal{H}_\mathrm{M} = - \frac{1}{\rho} \langle B' . \nabla \times B' \rangle_\phi.$$

## $\gamma$-effect
Magnetic pumping: $\mathbf{\gamma} = -\frac{\tau}{3}\nabla \langle {u'}^2\rangle_\phi$

\begin{equation}
\gamma_\mathrm{r} = -\frac{\tau}{3} \frac{d \langle {u'}^2\rangle_\phi}{dr} \propto  - \frac{d \langle {u'}^2\rangle_\phi}{dr}
\end{equation}

\begin{equation}
\gamma_\theta = -\frac{\tau}{3} \frac{d\langle {u'}^2\rangle_\phi}{r d\theta} \propto - \frac{d\langle {u'}^2\rangle_\phi}{r d\theta}
\end{equation}

```python
import numpy as N
import matplotlib.pyplot as P
import pandas as pd
P.style.use(['seaborn-white', 'seaborn-paper', 'seaborn-ticks'])
matplotlib.rc("font", family="Times New Roman", size=12)
matplotlib.rc('xtick', labelsize='medium')
matplotlib.rc('ytick', labelsize='medium')
```

```python
sims= pd.read_csv('./tables/data_meanfield.txt', sep=",")
sims['color'] = sims['eta'].replace(0.4,'k').replace(0.6,'r')
```

__In the next plots we try to access how the three quantities related to the $\alpha$ (kinetic and magnetic helicities) and $\gamma$ effect vary for our set of simulations.__ 
- All the quantities display a saturation for increasing Rossby number (Increasing Rayleigh).
- The magnetic helicity is negligible in our runs.

```python
P.scatter(sims['Ro'],  sims['gamma'],   c=sims['color'],  marker='o', s=80)
P.yscale('log')
P.ylim((1.e-1, 2.e6))
P.xlabel('Ro', fontsize='large')
P.ylabel(r'$|\gamma|$', fontsize='large')
```

```python
P.scatter(sims['Ro'],  sims['hk'],   c=sims['color'],  marker='o', s=80)
P.yscale('log')
P.ylim((1.e-1, 2.e6))
P.xlabel('Ro', fontsize='large')
P.ylabel(r'$\mathcal{H}_\mathrm{K}$', fontsize='large')
P.title(r'Root-mean-square $\mathcal{H}_\mathrm{K}$', fontsize='large')
```

```python
P.scatter(sims['Ro'],  sims['hm'],   c=sims['color'],  marker='o', s=80)
P.yscale('log')
P.ylim((1.e-1, 2.e6))
P.xlabel('Ro', fontsize='large')
P.ylabel(r'$\mathcal{H}_\mathrm{M}$', fontsize='large')
P.title(r'Root-mean-square $\mathcal{H}_\mathrm{M}$', fontsize='large')
```

Runs close to the onset of convection resulted in the highest mangnetic helicities, even though they represent less then $1\%$ of the total helicity:

```python
P.scatter(sims['Ro'],  sims['hm']/(sims['hk']+sims['hm']),   c=sims['color'],  marker='o', s=80)
P.yscale('log')
P.ylim((1.e-5, 0.01))
P.xlabel('Ro', fontsize='large')
P.ylabel(r'$\mathcal{H}_\mathrm{M} \; (\mathcal{H}_\mathrm{T})$', fontsize='large')
```

__We can check whether the $\alpha$ and $\gamma$ effects are comparable by taking their ratio.__ Our results show that both have the same order of magnitude.

```python
P.scatter(sims['Ro'],  sims['hk']/sims['gamma'],   c=sims['color'],  marker='o', s=80)
P.xlabel('Ro', fontsize='large')
P.ylabel(r'$\frac{\mathcal{H}_\mathrm{K}}{|\gamma|}$', fontsize=20, rotation=0, labelpad=20.)
P.show()
```
