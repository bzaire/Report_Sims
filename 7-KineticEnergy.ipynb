# Kinetic energy staturation at high-Rayleigh

We follow previous papers that study the high-Rayleigh regim in which a saturation was observed for the kinetic energy. 

__Wrap-up:__

- _Featerstone et al. (2016a)_: HD and non-rotating simulations; Density and Ra are varied.

- _Featerstone et al. (2016b)_: HD but rotating simulations; Only Rayleigh are varied.

- _Orverdahl et al. (2018)_: Varies Pr and Ra to probe the effect of the Prandtl number in the saturation. 
    - the cutoff Ra that defines the saturation regim varies with Pr (it decreases with lower Pr).
    - the saturation level of the kinetic energy increases as Pr is reduced.

## Results

```python
import numpy as N
import matplotlib.pyplot as P
import pandas as pd
P.style.use(['seaborn-white', 'seaborn-paper', 'seaborn-ticks'])
matplotlib.rc("font", family="Times New Roman", size=12)
matplotlib.rc('xtick', labelsize='medium')
matplotlib.rc('ytick', labelsize='medium')
```

```python
#DATA FROM PC AND FC RUNS
sims1= pd.read_csv('./tables/data_results_nrho1.txt', sep=",")
sims1['color'] = sims1['eta'].replace(0.4,'r').replace(0.6,'k')
sims1['symbol'] = sims1['Pr'].replace(1,'d')
sims1['LoFohm'] = N.sqrt(2*sims1['Emag'])/(sims1['OhmDiss']/sims1['BouPower'])**(0.5)

sims15= pd.read_csv('./tables/data_results_nrho15.txt', sep=",")
sims15['color'] = sims15['eta'].replace(0.4,'r').replace(0.6,'k')
sims15['symbol'] = sims15['Pr'].replace(1,'v')
sims15['LoFohm'] = N.sqrt(2*sims15['Emag'])/(sims15['OhmDiss']/sims15['BouPower'])**(0.5)

sims3= pd.read_csv('./tables/data_results_nrho3.txt', sep=",")
sims3['color'] = sims3['eta'].replace(0.4,'r').replace(0.6,'k')
sims3['symbol'] = sims3['Pr'].replace(1,'p')
sims3['LoFohm'] = N.sqrt(2*sims3['Emag'])/(sims3['OhmDiss']/sims3['BouPower'])**(0.5)
```

```python
fig, (ax1, ax2, ax3, ax4) = P.subplots(1,4, figsize=(24,4))
ax1.scatter(sims3['Rac'], sims3['Ekin'], c=sims3['color'], marker='p', s=80)
ax1.scatter(sims15['Rac'], sims15['Ekin'], c=sims15['color'], marker='v', s=80)
ax1.scatter(sims1['Rac'], sims1['Ekin'], c=sims1['color'], marker='d', s=80)
#ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_ylim((1.e4, 2.e7))
ax1.set_xlabel('Ra (Ra$_\mathrm{c}$)', fontsize='large')
ax1.set_ylabel(r'$E_\mathrm{K}$', fontsize='large')

ax2.scatter(sims3['Rac'], sims3['ViscDiss'], c=sims3['color'], marker='p', s=80)
ax2.scatter(sims15['Rac'], sims15['ViscDiss'], c=sims15['color'], marker='v', s=80)
ax2.scatter(sims1['Rac'], sims1['ViscDiss'], c=sims1['color'], marker='d', s=80)
ax2.set_yscale('log')
ax2.set_ylim((1.e7, 1.e12))
ax2.set_xlabel('Ra (Ra$_\mathrm{c}$)', fontsize='large')
ax2.set_ylabel(r'$ \sigma^2$', fontsize='large')
ax2.set_title('Viscous Dissipation', fontsize='large')


ax3.scatter(sims3['Rac'], sims3['OhmDiss'], c=sims3['color'], marker='p', s=80)
ax3.scatter(sims15['Rac'], sims15['OhmDiss'], c=sims15['color'], marker='v', s=80)
ax3.scatter(sims1['Rac'], sims1['OhmDiss'], c=sims1['color'], marker='d', s=80)
ax3.set_xlabel('Ra (Ra$_\mathrm{c}$)', fontsize='large')
ax3.set_yscale('log')
ax3.set_ylim((1.e7, 1.e12))
ax3.set_ylabel(r'$(\nabla \times B)^2$', fontsize='large')
ax3.set_title('Ohmic Dissipation', fontsize='large')

ax4.scatter(sims3['Rac'], sims3['BouPower'], c=sims3['color'], marker='p', s=80)
ax4.scatter(sims15['Rac'], sims15['BouPower'], c=sims15['color'], marker='v', s=80)
ax4.scatter(sims1['Rac'], sims1['BouPower'], c=sims1['color'], marker='d', s=80)
ax4.set_xlabel(r'Ra (Ra$_\mathrm{c}$)', fontsize='large')
ax4.set_ylabel(r'$\mathcal{P}$', fontsize='large')
ax4.set_title('Boyancy Power', fontsize='large')
ax4.set_yscale('log')
ax4.set_ylim((1.e7, 1.e12))
```

The figure above shows several quantities as a function of the Rayleigh number. We chose to express the later in terms of the critical Rayleigh, which determines the minimum value necessary to drive convection in the system, because this criticality varies with the density constrast and with the setup adopted for $\frac{d\bar{s}}{dr}$ (i.e., either fully convective or partially convective in this work). 

- Our set of simulations display the saturation behaviour preivously reported in the literature for HD simulations. 

- This find suggests that dissipative effects due to the magnetic field present the saturation behaviour for high-Rayleigh numbers (3rd panel).

$$\mathrm{Ra} = \frac{g_\mathrm{o} r_\mathrm{o}^4}{c_\mathrm{p}\kappa\nu} \left| \frac{d\bar{s}}{dr} \right|_{r_\mathrm{o}}, \ \ \ \ \mathrm{Pr} = \frac{\nu}{\kappa}, \ \ \ \ \mathrm{Pm} = \frac{\nu}{\eta}, \ \ \ \ \text{and}  \ \ \ \ \mathrm{Ek} = \frac{\nu}{\Omega r_\mathrm{o}^2} $$


\begin{equation}
Rm = \frac{u_\mathrm{rms} r_\mathrm{o}}{\eta}
\end{equation}
